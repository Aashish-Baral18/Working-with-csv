```python
#dtypes parameter
pd.read_csv('Filtered_PM_Data.csv',dtype={'PM10':'int'})
#here it is showing error because the pm10 column contains float values and pandas is being safe
#and trying to stop the loss of those values
#pd.read_csv('Filtered_PM_Data.csv').info()
```

```python
#handling dates
#important
#when you use read csv , by default all date columns are passed as strings
#so we vant use the date functionalities in those dates
pd.read_csv('Filtered_PM_Data.csv').info()
```

```python
#continue
# in above output we see date datatype as a object, we know in pandas object is string
# so to overcome that
pd.read_csv('Filtered_PM_Data.csv',parse_dates=['date']).info()
#esma vaxaina, kina vaena vaneko mero date ko format namilera raixa y-m-d format chaiyo ni ta dosti
```
